### Group by sets using the .groupby( ) method

- Ideal to group by column with duplocaet values to create segments. Example: Sector or Industry.

```python
import pandas as pd
```

```python
# import fortune 1000 and set index as rank
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')

fortune.head(3)
```

```python
# Group by sector
sectors = fortune.groupby('Sector')

# a group by object needs methods called on it, how many groupings are there? - 21
len(sectors)
```

```python
# to prove the above:
fortune['Sector'].nunique()
```

```python
# how many rows fall into each group?
sectors.size()
```

```python
# return first rows in each group and last - this is good for extracting sample rows
sectors.first()
sectors.last()
```

### Retreive a group with the .get_group( ) method

```python
# get the energy sector - returns every company info in the energy sector
sectors.get_group('Energy')
```

### Usng the max( ) method
- If used on string colum such as company, the max value will be the the closest to the end of the alphabet. The inverse is true if using min( ). Numbers are intuitive, the max is the largest, nim is the smallest.

- Using max or min with a group by and specifcy a numeric column will yeild the max or min value among the group.

```python
# example of using max in general which defaults to the first col because a col is not specififed.
sectors.max().head()

sectors.min().head()
```

```python
# apply sum() to all numeric columns
sectors.sum().head()
```

```python
# you can apply various numeirc methods like mean - rounding the dec values for this example.
round(sectors[['Revenue', 'Profits']].mean().head(), 2)
```

```python
# which sector has the most profits? get the top 3
sectors[['Profits']].sum().sort_values(by='Profits', ascending=False).head(3)
```

```python
# which sector has the most employees? Get the top 3
sectors[['Employees']].sum().sort_values(by='Employees', ascending=False).head(3)
```

### Grouping by multiple columns
- To do this, we just add a list into the group by statment: group_by( ['group1', 'group2'] )

```python
# add a group
sectors = fortune.groupby(['Sector', 'Industry'])
```

```python
# now we can see 2 groups in a hierarchy.
sectors.size()
```

```python
# the same methods can be applied but we will now get a multi index data frame
sectors.sum()
```

### The agg( ) method
- reset sectors to one column for demonstration
- agg( )  allows us to do muleiple agg methods on various colums by using a dictionary with key value pairs.
- The key is the column whereas the agg method is the value.

```python
sectors = fortune.groupby('Sector')
```

```python
# provide dictionay with key value pairs for agg methods. This ex sums the rev and profits witin each sector whie giving the av employees.
sectors.agg({"Revenue":"sum", 
             "Profits":"sum",
             "Employees": "mean"}).head()
```

```python
# use a list to apply multiple operations for size, sum and mean across secotrs as a summary
sectors.agg(["size", "sum", "mean"]).head()
```

### You can combine the two methods- dictionary and list to get specifc metrics for each sector.
- Example: get the sum and mean for Revenue, sum and mean for Profts and the sum of employees

```python
sectors.agg({"Revenue":["sum", "mean"],
             "Profits":["sum", "mean"],
             "Employees": "sum"}).head()
```

### Iterating through groups

Extract the companies with the highest profits.
- using max( ) will not produce the enotre rwo so we need to loop over.

```python
# create a new data frame using the DataFrame call and use the columns from the fortune variable intially created as the start.
df_for = pd.DataFrame(columns=fortune.columns)

# now the df is ready for appending the companies with the highest profits
df_for
```

```python
for sector, data in sectors:
    highest_rev_in_group = data.nlargest(1, "Revenue")
    
    # re-assign back to df_for var
    df_for = df_for.append(highest_rev_in_group)  
```

```python
# preview df_for now and see the enitre row for each company within that sector with the highest revenue.
# Note the data frame itself is not sorted with revenue descending, only the highest rev comapny within each sector.
df_for.head()
```

#### Do a city based analysis rather than sector.

```python
df_cities = fortune.groupby("Location")

# same steps as previous to create a new data frame that gets looped over and appended
df_for2 = pd.DataFrame(columns=fortune.columns)

df_for2
```

```python
# create the for loop to loop over cities
for city, data in df_cities:
    highest_rev_in_city = data.nlargest(1, "Revenue")
    
    # add to the df
    df_for2 = df_for2.append(highest_rev_in_city)
```
