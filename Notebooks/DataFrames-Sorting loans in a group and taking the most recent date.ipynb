### Getting max values within groups-
This can be achieved in many ways but here we will look at:
- Sorting within a group to take the row with the most recent target variable using a date.
- In this instance, I am looking to group loan id's with multiple values for letter date. I want the row with the most recent letter date sent.
- See groupby( ) documentation for all methods: <a>https://pandas.pydata.org/docs/reference/groupby.html</a>

```python
import pandas as pd
```

```python
loans = pd.read_csv('../datasets/loans.csv', parse_dates = ['letter date', 'loan date', 'maturity date'])
```

```python
loans.head(20)
#loans.info()
```

```python
loans2 = loans.sort_values(by=['loan id', 'letter date'], ascending=False).groupby('loan id').first()
```

```python
loans2.head(20)
```

```python
loans2.reset_index(inplace=True)
```

```python
loans2.head(10)
```

### Apply time delta to loan date

```python
# add 30 days and create a new column
loans2['Expected_Matrurity_30'] = loans2['loan date'] + pd.Timedelta(days=30)+pd.Timedelta(days=1)
loans2.head(7)
```

```python
# add 60 days and create a new column
loans2['Expected_Matrurity_60'] = loans2['loan date'] + pd.Timedelta(days=60)+pd.Timedelta(days=1)
loans2.head(7)
```
