### Dataframes continued: .sort_values( )
- Sort methods:  you have to use inplace=True for perm changes or reassign

- Rank methods

By default, null values are placed at the end. To see the nulls first use na_position = "first"

```python
import pandas as pd
```

```python
nba = pd.read_csv('nba.csv')
nba.head()
```

```python
# sort by name - the default is ascending and the entire df is sorted under the hierarchy of name
nba.sort_values('Name').head(3)
```

```python
# sort in descending order
nba.sort_values('Name', ascending=False).head(3)
```

```python
# sort by a floating point
nba.sort_values('Age').head()
```

```python
# get the top 3 earning players
nba.sort_values('Salary', ascending=False).head(3)
```

### Multiple column sort:

- Creates a hiearchy, in example below, Team is sorted ascending(by default) then Name ascending within the team. If ascending = False, reverse aplh order occurs witht the saem hierarchy applied.

- Use a list to specify multiple cols

```python
nba.sort_values(['Team', 'Name'])
```

```python
# what if we want ascending for Team and descenfing for Name?
nba.sort_values(['Team', 'Name'], ascending = [True, False])
```

### Sort with Index using .sort_index( )

```python
# first take the index out of order and make it inplace
nba.sort_values(['Number', 'Salary', 'Name'], inplace=True)
```

```python
# the index is out of whack but we can fix it easily
nba.head()
```

```python
# now sort by index
nba.sort_index()
```

```python
### Rest the  index, this assigns a new index ot the recently changed index above
nba.reset_index()
```

### Rank( ) method

- Can be used in conjuction with sort to establish rankings of values or within groups.

- In the case of ties, sequence is skipped. For example if the top salary is tied between 4 people, the rank will be 1,1,1,1,5 so the second person will show as 5. Thi scan be seen numerous times in this set.

```python
# re exceute and get rid of nulls rows + convert floats to int
nba = pd.read_csv('nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba.head()
```

#### Rank the salaries in the NBA wiht the hghest being 1

```python
# using this sytnax doesnt get us a rank that we want...it ranks in reverse order
nba['Salary'].rank()
```

```python
# to fix the above and assign to a new column in the set
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype('int')

nba.head()
```

```python
# lets prove the rank worsk by sorting by salary, you can see it does!!!
nba.sort_values("Salary", ascending = False)
```
