### SQL like Window Functions
Windows functions are great for doing any type or rabking or calculation on on sets aka groups of data. We can easily perform running totals,averages and statsitical calcs usng them. These mimic SQL window functions by working with group by but thre isnt really a set way to of a Pandas equavilent.

Source Cited: <a>https://towardsdatascience.com/sql-window-functions-in-python-pandas-data-science-dc7c7a63cbb4</a>

### Row Number( )
In SQL, the ROW NUMBER function wokrs with and over clause to partition a group using an order by stateent. This esstially means we have a group, we order the group(s)w we desire and then appply a row number based on the former two statments.

SQL example: ROW_NUMBER() OVER (PARTITION BY var1 ORDER BY var1Date DESC) as rowNum
- This produces an ordered set that shows the most recent date as number 1 going down to the latest. This is esentially a ranking of rows by date.

```python
# we can mimick this in PANDAS by using:
df['days_lookback'] = df.sort_values(['date'], ascending=False).groupby(['ticker']).cumcount() + 1
```

```python
# The result is a rwo number applied to a descending order within each "window" or ticker grouping. 
# The most recent tocker data is labeled as 1 with the first row having the highest row number.
df
```

```python
# if we only want to get the most recent for each ticker, we can simply filter it!
df[df['days_lookback'] == 1]
```

### Percentile rank within each group
This is the equavilent of the percent rank which tells us the percentage that each price is within that group.
Example: FB has two perc_price with 1.0 but they both represent different years. You can see that for 2020 there are
three rows of data with each repreenting a percentage of the whole year. 1.0 represents the high price with 66.0 being 2nd and so on. 

```python
df['perc_price'] = df.groupby(['ticker', 'year'])['price'].rank(pct=True)
```

### Running sum  within groups

```python
df_aggs['volume_3day'] = df_aggs.sort_values(by=['date'], ascending=True)\
                       .groupby(['ticker'])['trade_vol']\
                       .rolling(3, min_periods = 1).sum()\
                       .reset_index(drop=True, level=0)
```

```python
df_aggs['cum_total_vol'] = df_aggs.sort_values(by=['date'], ascending=True)\
                       .groupby(['ticker'])['trade_vol']\
                       .cumsum()
```

```python
df_aggs['volume_3day'] = df_aggs.sort_values(by=['date'], ascending=True)\
                       .groupby(['ticker'])['trade_vol']\
                       .rolling(3, min_periods = 1).sum()\
                       .reset_index(drop=True, level=0)
```
