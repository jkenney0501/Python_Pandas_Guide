### Merge, Join & Concatenate DataFrames
- This section will use a series of different notebooks for restaraunt data where we can concatenate, merge and join table sbased on their relationship via keys similar to the way we would join tables in SQL using a primary/foreign key relationship.
- This section is also fundamental for any Data Analyst role.

#### Concatenating
- To Concatenate to to add to data frames toghether on top of one another. We need to have to the same columns names for this otherwise a new column will be added that doesnt belong to one of the data frames.

```python
# import data sets 

# week 1 and 2 have the same cols so we can concatentate them bc they are just different time periods.
# the other two will require merges or joins based on key relationships.
week1 = pd.read_csv('../datasets/Restaurant - Week 1 Sales.csv')

week2 = pd.read_csv('../datasets/Restaurant - Week 2 Sales.csv')

customers = pd.read_csv('../datasets/Restaurant - Customers.csv')

foods = pd.read_csv('../datasets/Restaurant - Foods.csv')

week1.head(2)
week2.head(2)
customers.head(2)
foods.head(2)
```

#### Lets concatenate week1 and week 2 given they have the same structure. Take a look at the function paramters by hitting shift tab.

- Simply use objs = [add a list of df's to concatenate!]
- The order in which they are passed is the order the appear.
- Note the index is added on as well, it is not reset from 1-500, its 1-250 for week1 and starts again at week2 1-250.
- To get a new index use ignore_index=True


```python
# you can assign this to a new varibale to beging your work but for demo purposes I am leaving it as is.

# Index is concatenated:
pd.concat(objs = [week1, week2])

# Index is reset
pd.concat(objs = [week1, week2], ignore_index=True)
```

#### You can acheive the same results with the append( ) method.
- Rather than objs= we use other = as a paramter. 

```python
week1.append(other=week2, ignore_index=True)
```

#### Add a unique index while keeping the original.
- Common someties when working with survey data.

```python
# This creates a multi index DF
pd.concat(objs = [week1, week2], keys=['week1', 'week2'])
```

```python
# Assign to a varibale and extarct using index position/loc
sales = pd.concat(objs = [week1, week2], keys=['week1', 'week2'])

sales.loc[('week1')].head()
```

```python
# get position 240
sales.loc[('week1', 240)]
```

```python
# access multiple cols
sales.loc[('week2', 183), ['Customer ID', 'Food ID']]
```

#### Inner Joins - merge( ) or pd.merge( ) is used for all types of joins
- Merges data sets based on common shared values using the merge( ) method with the parameter how = "inner". Shift-tab to see more.
- Example: We have 2 data sets, employees and sales. Not every employee is responsonsible for sales and if we merged with a condition of at least 1 sale, we would only get matching reults for those employees with a sale.
- Think of a Venn Diagram, this is the inner section where both events occur. As above, an inner join produces a joined dataset where an employee is in sales and has a sale.

```python
# we are calling the left data frame (week1) and merging the right data frame to it, the default is inner.
# the "on" paramter requires both have the same name.
# notice the new col names with suffixes of _x (week1), _y(week2) this is to show what cols matched in both without having dup cols.
# also notice, the inner mrge doesnt de dup, look at customer id 155, week1 had 2 results that are different but only 1 in week 2.
week1.merge(week2, how = 'inner', on = 'Customer ID').head()
```

#### Change the default suffixes by using the suffixes =  parameter.

```python
week1.merge(week2, how = 'inner', on = 'Customer ID', suffixes=('_w1', '_w2')).head()
```

#### Inner join on multiple columns
- Note the diference: there arent multiple week columns becasue previously we only merged on a Customer ID and now we use both so essentially we are passing a condition that must be met and can be rendered with the same columns to represent that condition. The previous condition rendered different results in both weeks.

```python
# find the customers that came in both weeks and ordered the exact same food item
# for this pass a list into the on paramter. Cols must have the same name.
week1.merge(week2, how = 'inner', on = ['Customer ID', 'Food ID'])
```

#### Outer Joins
- Includes only rows that belong in either. It is the opposite of the inner join.
- Think of the Venn Diagram where one is A and the other is B, the innner join is the instersection or $A \cap B$ whereads the outer join is A or B AKA A union B: $ A\cup B$

```python
# The syntaxt is very much the saem with the change in the how = "outer"
# Notice we now have nulls (NaN)? An outer requests either Left or right results, if they dont happen in both, we get null.
# Elaboration: where not null means rows are found in both tables, where null means it doesnt exist in that table.
# Customer ID 658 shows that in week 1 we had food id 1 but not in week 2. 
# Notice the pandas default to a float due to nulls. This can be fixed by replacing nulls with 0 then converting to INT
week1.merge(week2, how = 'outer', on = 'Customer ID', suffixes=('_w1', '_w2'))
```

```python
# we can use indicator = True to see where the dat is being pulled from - this is really handy for dev work!
week1.merge(week2, how = 'outer', on = 'Customer ID', suffixes=('_w1', '_w2'), indicator=True)
```

```python
# look at the value distribution
merged = week1.merge(week2, how = 'outer', on = 'Customer ID', suffixes=('_w1', '_w2'), indicator=True)

merged['_merge'].value_counts()
```

#### Using isin( ) to find values from the outer join.
- This can be helpful for identifying a number of questions.

```python
# to see the rows, wrap the query
merged[merged['_merge'].isin(['both'])].head()
```

#### Left Joins
- also covered: left on and right on paramters
- A left join takes all values from the left table and joins the right table showing all values that macth and null where they do not.
- Think of the Sales example. If we use employees table as left and sales as the right, not all employees have sales becaseu they arent in sales. We will still see those rows but null values for sales.
- Note: a right join is the saem output but the right table is the base table or foundation so it will not be performed here.

```python
# First get counts to conceptuallize a left
week1.count() # 250
foods.count() # 10

# 250 on the left table and 10 on the right. Technically, all 10 values will be used but they dont have to be.
```

```python
# using week 1 and foods, we can see the common key is food id so lest left join it
# usng pd.merge method
left = pd.merge(week1, foods, how = 'left', on = 'Food ID')

left.head()
```

#### Left on and right on paramters
- Used when they keys have different names but represent the same values. Example: one table might be customer id and another may call the same key client id. In tis situation, we use left and right on paramters.
- A good example is the customers table and week1. The common key is customer ID(week1) and ID(customers.

```python
pd.merge(week1, customers, how = 'left', left_on = 'Customer ID', right_on='ID').head()
```

```python
# drop the extra ID becaseu we dont need it twice by using drop() - use axis = columns or 1
pd.merge(week1, customers, how = 'left', left_on = 'Customer ID', right_on='ID').drop('ID', axis=1)
```

#### Merge with index
- We have been using it but heres a little more info.
- and you can merge with the index
- Advantage is the additional index col isnt listed as column we need to drop.

```python
# re import and set the customers ID as the index
# we are going to merge to the left table with the right index
week1 = pd.read_csv('../datasets/Restaurant - Week 1 Sales.csv')
foods = pd.read_csv('../datasets/Restaurant - Foods.csv', index_col = 'Food ID')
customers = pd.read_csv('../datasets/Restaurant - Customers.csv', index_col= 'ID')
```

```python
# ID is now the index we ill merge to week1 customer ID
customers.head(2)
```

```python
# set the right index = True, this tells pandas we are using it as a key to join on.
week1.merge(customers, how = 'left', left_on = 'Customer ID', right_index = True)
```
